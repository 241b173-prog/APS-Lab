# Shortest Path Using Dijkstra Algorithm in Python
Problem Description:
Given a weighted graph with non negative edge weights. Find the shortest distance from a source vertex to all other vertices using Dijkstra algorithm

#Theory
Dijkstra algorithm finds shortest distance from one source node to all other nodes in a weighted graph with non negative weights. Dictionary structure stores nodes as keys and neighbor nodes with weights as values.

##Algorithm Steps

1. Represent graph using dictionary.
2. Set distance of every node as infinity.
3. Set distance of source node as 0.
4. Select node with smallest distance among unvisited nodes.
5. Update distance of neighbor nodes.
6. Repeat until all nodes get visited.

```python
def dijkstra(graph, start):

    distances = {}
    visited = []

    for node in graph:
        distances[node] = float('inf')

    distances[start] = 0

    while len(visited) < len(graph):

        min_node = None
        min_distance = float('inf')

        for node in graph:
            if node not in visited and distances[node] < min_distance:
                min_distance = distances[node]
                min_node = node

        visited.append(min_node)

        for neighbor, weight in graph[min_node].items():
            new_distance = distances[min_node] + weight

            if new_distance < distances[neighbor]:
                distances[neighbor] = new_distance

    return distances
```

```python
graph = {
'A': {'B': 4, 'C': 2},
'B': {'A': 4, 'C': 5, 'D': 10},
'C': {'A': 2, 'B': 5, 'D': 3},
'D': {'B': 10, 'C': 3}
}

result = dijkstra(graph, 'A')

for node in result:
    print("Distance from A to", node, "=", result[node])
```

```python
def dijkstra(graph, start):
    distances = {node: float('inf') for node in graph}
    distances[start] = 0
    visited = []

    while len(visited) < len(graph):
        min_node = None
        min_distance = float('inf')

        for node in graph:
            if node not in visited and distances[node] < min_distance:
                min_distance = distances[node]
                min_node = node

        if min_node is None:  # safeguard if graph is disconnected
            break

        visited.append(min_node)

        for neighbor, weight in graph[min_node].items():
            new_distance = distances[min_node] + weight
            if new_distance < distances[neighbor]:
                distances[neighbor] = new_distance

    return distances


# -------------------------------
# User Input Graph
# -------------------------------
graph = {}
n = int(input("Enter number of nodes: "))

for i in range(n):
    node = input(f"Enter node {i+1} name: ").strip()
    graph[node] = {}
    neighbors = int(input(f"Enter number of neighbors for {node}: "))
    for j in range(neighbors):
        neighbor = input(f"Neighbor {j+1} name: ").strip()
        weight = int(input(f"Weight from {node} to {neighbor}: "))
        graph[node][neighbor] = weight

start = input("Enter source node: ").strip()

result = dijkstra(graph, start)

print("\nShortest distances:")
for node in result:
    print(f"Distance from {start} to {node} = {result[node]}")
```

```python
def dijkstra_with_path(graph, start):
    distances = {node: float('inf') for node in graph}
    distances[start] = 0
    visited = []
    predecessors = {node: None for node in graph}

    while len(visited) < len(graph):
        min_node = None
        min_distance = float('inf')

        for node in graph:
            if node not in visited and distances[node] < min_distance:
                min_distance = distances[node]
                min_node = node

        visited.append(min_node)

        for neighbor, weight in graph[min_node].items():
            new_distance = distances[min_node] + weight
            if new_distance < distances[neighbor]:
                distances[neighbor] = new_distance
                predecessors[neighbor] = min_node

    return distances, predecessors

def get_path(predecessors, node):
    path = []
    while node is not None:
        path.insert(0, node)
        node = predecessors[node]
    return path

graph = {
    'A': {'B': 4, 'C': 2},
    'B': {'A': 4, 'C': 5, 'D': 10},
    'C': {'A': 2, 'B': 5, 'D': 3},
    'D': {'B': 10, 'C': 3}
}

start = 'A'
distances, predecessors = dijkstra_with_path(graph, start)

for node in graph:
    path = get_path(predecessors, node)
    print(f"Shortest path from {start} to {node}: {' -> '.join(path)} with distance {distances[node]}")
```
